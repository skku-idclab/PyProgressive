```python
import pyprogressive as pp
from pyprogressive import each, accum
```

```python
array = pp.Array([10, 20, 0, 21, 5, 42, 11, 14, 34, 13])

avg = accum(each(array))/len(array)
var = accum((each(array) - avg)**2) / len(array)

compiled = pp.compile(avg, var)
compiled.run(callback = lambda avg, var: print(avg, var))
```
